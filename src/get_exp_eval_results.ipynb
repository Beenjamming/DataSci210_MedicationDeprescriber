# Evaluate Experiment Results

Use this script to evaluate the results of experiments that are populated in the folder Data/Experiments.

```python
import pandas as pd
import numpy as np
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
import os
import json
from pathlib import Path
import pprint
```

```python
from dotenv import load_dotenv
load_dotenv()

groq_key = os.getenv("groq_key_paid")

data_path = os.getenv("data_path")
data_path = Path(data_path)
```

```python
from eval import evaluate_multiclass_classification, get_best_diagnosis_match
```

```python
def get_results(name):
    experiment_df = pd.read_csv(data_path / "Experiments" / f"{name}.csv")
    
    y_true = experiment_df['recommendation']
    y_pred = experiment_df['ret_recommendation']
    class_labels = ['continue', 'deprescribe', 'stop']

    # Evaluate the model
    metrics = evaluate_multiclass_classification(y_true, y_pred, class_labels)
    return metrics
```
